```python
input = np.array([2, 3, 4])

# 2. Combine weights into a single 2D matrix W (shape: 3 neurons x 3 inputs)
Weights = np.array([
    [0.1, 0.2, 0.3],  # weights1
    [0.2, 0.3, 0.4],  # weights2
    [0.3, 0.4, 0.5]   # weights3
])

# 3. Combine biases into a single 1D array b
bias = np.array([-5, 6, 2])

# 4. Compute output using dot product
output = np.dot(Weights, input) + bias
output=np.maximum(0,output)
print(output.tolist())
```

```python
inputs = [2, 3, 4, 5]


weights = [
    [0.1, 0.2, 0.3],  # Neuron 1 weights
    [0.2, 0.3, 0.4],  # Neuron 2 weights
    [0.3, 0.4, 0.5]   # Neuron 3 weights
]

# 3. Combined bias vector (List)
bias = [-5, 6, 2]

# 4. Compute output using loops
output = []
for i in range(len(weights)):
    neuron_sum = 0
    for j in range(len(weights[i])):
        neuron_sum += inputs[j] * weights[i][j]
    
    # Add the bias for the current neuron and save the result
    output.append(neuron_sum + bias[i])
output=np.maximum(0,output)
print(output)
```

```python
inputs = [[0.3, 0.5, -0.3, 2],
          [0.6, -0.83, 0.59, -0.6],
          [6, 1, 7, 1]]

weights = [
     [0.1, 0.2, 0.3, 0.4],
     [0.2, 0.3, 0.4, 0.5],
     [0.3, 0.4, 0.5, 0.6]
]

bias = [-5, 6, 2]

# Compute linear combination layer outputs
layer_output = np.dot(inputs, np.array(weights).T) + bias

# Define custom sigmoid activation function 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Apply activation to the matrix
activated_output = sigmoid(layer_output)

print(activated_output)
```

```python

# 1. Inputs: 3 students, 4 features each
# Features: [Attendance, Midterm, Assignments, Final Exam]
inputs = np.array([
    [0.95, 0.85, 0.90, 0.88],  # Student 1
    [0.40, 0.50, 0.45, 0.30],  # Student 2
    [0.75, 0.65, 0.70, 0.60]   # Student 3
])

# 2. Weights: 1 row representing 1 output neuron (Pass/Fail)
weights = np.array([
    [2.0, 1.5, 1.5, 4.0]
])

bias = np.array([-4.5])

output = np.dot(inputs, weights.T) + bias

# 5. Activation Function: Sigmoid converts scores to probabilities between 0 and 1
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

probabilities = sigmoid(output)

# 6. Classification: 1 if Probability >= 0.5 (Pass), else 0 (Fail)
predictions = (probabilities >= 0.5).astype(int)

# Display Results
for i, (prob, pred) in enumerate(zip(probabilities.flatten(), predictions.flatten())):
    status = "Pass" if pred == 1 else "Fail"
    print(f"Student {i+1}: {status}")
```
